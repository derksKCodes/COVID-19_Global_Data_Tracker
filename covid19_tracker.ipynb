```python
# For data manipulation
import pandas as pd

# For numerical operations (optional but useful)
import numpy as np

import seaborn as sns
# For visualizations (used in later steps, but good to import now)
import matplotlib.pyplot as plt
# Set styles for plots (optional for better visuals)
sns.set(style="whitegrid")
```

```python
# STEP 2: Data loading and Exploration
df = pd.read_csv("owid-covid-data.csv")
display(df)
display(df.info())
print(f"Dataset conatains {df.shape[0]:} rows and {df.shape[1]:} columns\n")
print("Column names:\n", df.columns.tolist(), "\n")
print("First 5 rows of the dataset:")
display(df.head())
print("\nMissing Values in each column:")
display(df.isnull().sum().sort_values(ascending=False))

display(df.size)
```

```python
#STEP 3: Data Cleaning
countries = ["Kenya","United States","India"]
df = df[df["location"].isin(countries)]
print("\nSummary of missing values before cleaning:")
display(df.isnull().sum())
df=df.dropna(subset=["date", "total_cases"])
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values(['location', 'date'])
display(df.info())

df.fillna(method='ffill',inplace=True)
df.fillna(0, inplace=True)
display(df.head())
display(df.size)

print("\nSummary of missing bvalues after cleaning:")
display(df.isnull().sum())
```

```python
#STEP 4: Exploratory Data Analysis
df.describe()

#. Plot Total Cases Over Time (Line Chart)
plt.figure(figsize=(13, 6))
for country in ['Kenya', 'United States', 'India']:
    country_data = df[df['location']==country]
    plt.plot(country_data['date'], country_data['total_cases'], label=country)
plt.title('\n\nTotal COVID-19 cases Ove Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# Plot Total Deaths Over Time (Line Chart)
plt.figure(figsize=(12, 6))
for country in ['Kenya', 'United States', 'India']:
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], country_data['total_deaths'], label=country)

plt.title('\n\nTotal COVID-19 Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# Daily New Cases Comparison
plt.figure(figsize=(12, 6))
for country in ['Kenya', 'United States', 'India']:
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], country_data['new_cases'], label=country)

plt.title('\n\nDaily New COVID-19 Cases')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


#Bar Chart: Death Rate by Country
# Get the most recent record for each country
latest_df = df.groupby('location').last().reset_index()

# Avoid divide-by-zero by filtering where total_cases > 0
latest_df = latest_df[latest_df['total_cases'] > 0]

# Calculate death rate
latest_df['death_rate'] = latest_df['total_deaths'] / latest_df['total_cases']

# Sort by death rate (optional)
latest_df = latest_df.sort_values(by='death_rate', ascending=False)

# Plot
plt.figure(figsize=(15, 8))
sns.barplot(x='death_rate', y='location', data=latest_df, hue='location', palette='Reds_r', legend=False)
plt.title('\n\nCOVID-19 Death Rate by Country')
plt.xlabel('Death Rate')
plt.ylabel('Country')
plt.tight_layout()
plt.show()


#Heatmap for Correlation Analysis
# Select numeric columns
numeric_df = df[['total_cases', 'new_cases', 'total_deaths', 'new_deaths', 'total_vaccinations']].copy()
numeric_df = numeric_df.dropna()

# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between COVID-19 Metrics')
plt.tight_layout()
plt.show()
```
